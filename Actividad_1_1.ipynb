## PRIMER FILTRO

En el primer filtro decidí hacerlo por el nombre del cliente, señalando aquellos movimientos realizados por alguna compra del Instituto Politecnico Nacional

```python
#Filtro por objeto 
filtro1=df[df["NOMBRE_CLIENTE"] == "INSTITUTO POLITECNICO NACIONAL"]
filtro1
```

```python
#Convertir archivo filtrado a CSV
filtro1.to_csv("Filtro1.csv")
```

```python
#descargar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Filtro1.csv")
```

## Segundo filtro: filtro por filas

En el segundo filtro, aquél por filas, se aplicaron desde la primera compra registrada por el Instituto, hasta la última vez. Sin embargo, se conservaron demás ventas registradas pues se hizo el análisis con filas consecutivas. 

```python
#Filtro por filas
filtro2= df.iloc[58:1332,: ]           #Filas consecutivas

filtro2
```

```python
#Convertir archivo filtrado a CSV
filtro2.to_csv("Filtro2.csv")
```

```python
#descargar archivo filtrado en csv
from google.colab import files 
  
  
files.download("Filtro2.csv")
```

## Filtro 3: por columnas
